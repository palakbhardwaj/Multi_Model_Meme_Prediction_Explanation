```python
import pandas as pd
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import os
```

```python
df = pd.read_csv("dataset/memotion_dataset_7k/labels.csv")

# clean
df = df.drop(columns=["Unnamed: 0"])

# small subset (important for laptop)
df = df.sample(100, random_state=42).reset_index(drop=True)

# use clean text
df['text'] = df['text_corrected'].fillna("")
```

```python
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

```python
labels = ["positive meme", "negative meme"]
```

```python
def predict_multimodal(row):
    try:
        image_path = os.path.join(
            "dataset/memotion_dataset_7k/images",
            row['image_name']
        )
        
        text = row['text']
        image = Image.open(image_path).convert("RGB")
        
        inputs = processor(
            text=[text + " " + l for l in labels],
            images=image,
            return_tensors="pt",
            padding=True
        )
        
        outputs = model(**inputs)
        probs = outputs.logits_per_image.softmax(dim=1)
        
        idx = torch.argmax(probs).item()
        
        label = "positive" if idx == 0 else "negative"
        score = probs[0][idx].item()
        
        return label, score
    
    except:
        return "error", 0.0
```

```python
results = []

for _, row in df.iterrows():
    pred, score = predict_multimodal(row)
    
    results.append({
        "image": row['image_name'],
        "predicted": pred,
        "confidence": score,
        "actual": row['overall_sentiment']
    })

result_df = pd.DataFrame(results)
result_df.head()
```

```python
def explain(row):
    text = row['text']
    pred = row['predicted']
    
    humour = row['humour']
    sarcasm = row['sarcasm']
    offensive = row['offensive']
    
    explanation = f"Meme: '{text[:40]}'. "
    
    if pred == "positive":
        explanation += "Positive because "
        
        reasons = []
        
        if humour in ["funny", "very_funny", "hilarious"]:
            reasons.append("it is humorous")
        if sarcasm == "not_sarcastic":
            reasons.append("it has no sarcasm")
        if offensive == "not_offensive":
            reasons.append("it is not offensive")
        
        if not reasons:
            reasons.append("the tone is pleasant")
        
        explanation += ", ".join(reasons) + "."
    
    elif pred == "negative":
        explanation += "Negative because "
        
        reasons = []
        
        if sarcasm != "not_sarcastic":
            reasons.append("it contains sarcasm")
        if offensive != "not_offensive":
            reasons.append("it may be offensive")
        if humour == "not_funny":
            reasons.append("it lacks humor")
        
        if not reasons:
            reasons.append("the tone is negative")
        
        explanation += ", ".join(reasons) + "."
    
    return explanation
```

```python
df["predicted"] = result_df["predicted"]

df["explanation"] = df.apply(explain, axis=1)

df[["image_name", "predicted", "explanation"]].head()
```

```python
import pandas as pd

df = pd.read_csv("dataset/memotion_dataset_7k/labels.csv")

df = df.drop(columns=["Unnamed: 0"])

# small subset
df = df.sample(100, random_state=42).reset_index(drop=True)

df['text'] = df['text_corrected'].fillna("")
```

```python
from transformers import pipeline

sentiment_model = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
def predict_sentiment(text):
    result = sentiment_model(text[:512])[0]
    
    label = "positive" if result['label'] == "POSITIVE" else "negative"
    score = result['score']
    
    return label, score
    
```

```python
results = []

for _, row in df.iterrows():
    label, score = predict_sentiment(row['text'])
    
    results.append({
        "predicted": label,
        "confidence": score
    })

result_df = pd.DataFrame(results)

df["predicted"] = result_df["predicted"]
```

```python
from transformers import pipeline, CLIPProcessor, CLIPModel
from PIL import Image
import torch, os

# text model
text_model = pipeline("sentiment-analysis")

# image model
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

labels = ["positive meme", "negative meme"]
```

```python
def text_sentiment(text):
    result = text_model(text[:512])[0]
    label = 1 if result['label'] == "POSITIVE" else 0
    return label, result['score']
```

```python
def image_sentiment(row):
    path = os.path.join("dataset/memotion_dataset_7k/images", row['image_name'])
    
    image = Image.open(path).convert("RGB")
    
    inputs = clip_processor(
        text=labels,
        images=image,
        return_tensors="pt",
        padding=True
    )
    
    outputs = clip_model(**inputs)
    probs = outputs.logits_per_image.softmax(dim=1)
    
    idx = torch.argmax(probs).item()
    
    return idx, probs[0][idx].item()
```

```python
def multimodal_sentiment(row):
    t_label, t_score = text_sentiment(row['text'])
    i_label, i_score = image_sentiment(row)
    
    # weighted fusion
    final_score = (t_label * t_score + i_label * i_score) / (t_score + i_score + 1e-6)
    
    final_label = "positive" if final_score >= 0.5 else "negative"
    
    return final_label
```

```python
df = df.sample(50)

df["predicted"] = df.apply(multimodal_sentiment, axis=1)
```

```python
def explain(row):
    text = row['text']
    pred = row['predicted']
    
    humour = str(row['humour']).lower()
    sarcasm = str(row['sarcasm']).lower()
    offensive = str(row['offensive']).lower()
    
    explanation = f"This meme is {pred} because "
    
    reasons = []
    
    # text-based reasoning
    if pred == "positive":
        if "funny" in humour:
            reasons.append("the text is humorous")
        if "not" in offensive:
            reasons.append("it is not offensive")
    
    else:
        if "sarcasm" in sarcasm:
            reasons.append("the text contains sarcasm")
        if "offensive" in offensive:
            reasons.append("it may be offensive")
    
    # image reasoning
    reasons.append("the image context supports this sentiment")
    
    return explanation + ", ".join(reasons) + "."
```

```python
df_30 = df.head(30)

df_30["explanation"] = df_30.apply(explain, axis=1)

df_30[["image_name", "predicted", "explanation"]]
```

```python
from sentence_transformers import SentenceTransformer, util

# load lightweight BERT model
bert_model = SentenceTransformer('all-MiniLM-L6-v2')

def semantic_similarity(row):
    text = row['text']
    explanation = row['explanation']
    
    emb1 = bert_model.encode(text, convert_to_tensor=True)
    emb2 = bert_model.encode(explanation, convert_to_tensor=True)
    
    score = util.cos_sim(emb1, emb2).item()
    
    return score
```

```python
from transformers import pipeline

nli_model = pipeline("text-classification", model="facebook/bart-large-mnli")

def nli_score(row):
    text = row['text']
    explanation = row['explanation']
    
    result = nli_model(f"{text} </s> {explanation}")[0]
    
    return result['score'] if result['label']=="ENTAILMENT" else 0
```

```python
df_30["nli_score"] = df_30.apply(nli_score, axis=1)
```
